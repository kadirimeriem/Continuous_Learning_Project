## Introduction
Le cycle de vie d’un modèle d’IA est composé de plusieurs étapes, de la collecte des données à la mise en production. fois entraıné et évalué, un modéle d’IA doit etre déployé pour etre utilisé dans des applications réelles. Il doit également etre maintenu et mis à jour régulierement. Selon le rythme des mises à jour et la quantité des données utilisées pour cela, on distingue deux grandes approches :
 
 — l’apprentissage continu (ou incréemental) : les modèles sont mis a jour a intervalle régulier en utilisant des batchs de taille assez grande.  
 
 — l’apprentissage en ligne (ou stream learning) : les modèles sont mis àjour en temps réel a chaque nouvelle donnée.
Dans ce projet, nous allons nous intéresser a la conception et la mise en oeuvre de solutions pour ces deux approches. Dans la suite, nous utiliserons le terme d’apprentissage continu pour désigner les deux approches. La distinction se fera ensuite sur la taille des batchs utilisés pour la mise à jour des modèles.

### Premier Objectif
Le premier objectif de ce projet est de comparer les performances des deux méthodes d’apprentissage continu évoquées ci-dessus. On va comparer les
performances en termes de temps de calcul, de mémoire utilisée, de qualité des modèles obtenus, etc. 

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, Input
import time
import tracemalloc
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load and preprocess data
def load_and_preprocess_data():
    (mnist_x_train, mnist_y_train), (mnist_x_test, mnist_y_test) = tf.keras.datasets.mnist.load_data()
    (fmnist_x_train, fmnist_y_train), (fmnist_x_test, fmnist_y_test) = tf.keras.datasets.fashion_mnist.load_data()

    mnist_x_train = mnist_x_train[..., np.newaxis].astype("float32") / 255.0
    mnist_x_test = mnist_x_test[..., np.newaxis].astype("float32") / 255.0
    fmnist_x_train = fmnist_x_train[..., np.newaxis].astype("float32") / 255.0
    fmnist_x_test = fmnist_x_test[..., np.newaxis].astype("float32") / 255.0

    fmnist_y_train = fmnist_y_train + 10
    fmnist_y_test = fmnist_y_test + 10

    return (mnist_x_train, mnist_y_train, mnist_x_test, mnist_y_test,
            fmnist_x_train, fmnist_y_train, fmnist_x_test, fmnist_y_test)

# Create base CNN model
def create_base_model():
    inputs = Input(shape=(28, 28, 1))
    x = layers.Conv2D(32, (3, 3), activation='relu')(inputs)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Conv2D(64, (3, 3), activation='relu')(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Conv2D(64, (3, 3), activation='relu')(x)
    x = layers.Flatten()(x)
    x = layers.Dense(64, activation='relu')(x)
    outputs = layers.Dense(20, activation='softmax')(x)
    model = models.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Function to create initial training set
def create_initial_dataset(mnist_x_train, mnist_y_train, fmnist_x_train, fmnist_y_train, initial_size=0.2):
    mnist_indices = np.random.choice(len(mnist_x_train), size=int(len(mnist_x_train) * initial_size), replace=False)
    fmnist_indices = np.random.choice(len(fmnist_x_train), size=int(len(fmnist_x_train) * initial_size), replace=False)
    
    x_initial = np.concatenate([mnist_x_train[mnist_indices], fmnist_x_train[fmnist_indices]])
    y_initial = np.concatenate([mnist_y_train[mnist_indices], fmnist_y_train[fmnist_indices]])

    remaining_mnist_x = np.delete(mnist_x_train, mnist_indices, axis=0)
    remaining_mnist_y = np.delete(mnist_y_train, mnist_indices, axis=0)
    remaining_fmnist_x = np.delete(fmnist_x_train, fmnist_indices, axis=0)
    remaining_fmnist_y = np.delete(fmnist_y_train, fmnist_indices, axis=0)
    
    x_remaining = np.concatenate([remaining_mnist_x, remaining_fmnist_x])
    y_remaining = np.concatenate([remaining_mnist_y, remaining_fmnist_y])
    
    return (x_initial, y_initial), (x_remaining, y_remaining)

# Implement Brute Force (BF) approach
def brute_force_update(model, new_batch_x, new_batch_y, all_previous_x, all_previous_y):
    tracemalloc.start()
    start_time = time.time()
    x_combined = np.concatenate([all_previous_x, new_batch_x])
    y_combined = np.concatenate([all_previous_y, new_batch_y])
    history = model.fit(x_combined, y_combined, epochs=1, batch_size=500, verbose=0)
    training_time = time.time() - start_time
    _, peak_memory = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return model, training_time, peak_memory / 1e6, history.history['accuracy'][-1]

# Implement Naive Fine-Tuning (NFT) approach
def naive_fine_tuning(model, new_batch_x, new_batch_y):
    tracemalloc.start()
    start_time = time.time()
    history = model.fit(new_batch_x, new_batch_y, epochs=1, batch_size=500, verbose=0)
    training_time = time.time() - start_time
    _, peak_memory = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return model, training_time, peak_memory / 1e6, history.history['accuracy'][-1]

# Main experiment function
def run_experiment():
    (mnist_x_train, mnist_y_train, mnist_x_test, mnist_y_test,
     fmnist_x_train, fmnist_y_train, fmnist_x_test, fmnist_y_test) = load_and_preprocess_data()
    
    (x_initial, y_initial), (x_remaining, y_remaining) = create_initial_dataset(
        mnist_x_train, mnist_y_train, fmnist_x_train, fmnist_y_train
    )

    base_model = create_base_model()
    base_model.fit(x_initial, y_initial, epochs=5, batch_size=500, validation_split=0.2)

    n_batches = 10
    batch_size = len(x_remaining) // n_batches

    bf_times = []
    nft_times = []
    bf_memories = []
    nft_memories = []
    bf_accuracies = []
    nft_accuracies = []

    x_test = np.concatenate([mnist_x_test, fmnist_x_test])
    y_test = np.concatenate([mnist_y_test, fmnist_y_test])

    bf_model = tf.keras.models.clone_model(base_model)
    bf_model.set_weights(base_model.get_weights())
    nft_model = tf.keras.models.clone_model(base_model)
    nft_model.set_weights(base_model.get_weights())
    
    bf_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    nft_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    prev_x = x_initial.copy()
    prev_y = y_initial.copy()

    for i in range(n_batches):
        batch_start = i * batch_size
        batch_end = (i + 1) * batch_size
        batch_x = x_remaining[batch_start:batch_end]
        batch_y = y_remaining[batch_start:batch_end]

        bf_model, bf_time, bf_memory, bf_acc = brute_force_update(bf_model, batch_x, batch_y, prev_x, prev_y)
        prev_x = np.concatenate([prev_x, batch_x])
        prev_y = np.concatenate([prev_y, batch_y])

        nft_model, nft_time, nft_memory, nft_acc = naive_fine_tuning(nft_model, batch_x, batch_y)

        bf_times.append(bf_time)
        nft_times.append(nft_time)
        bf_memories.append(bf_memory)
        nft_memories.append(nft_memory)
        bf_accuracies.append(bf_acc)
        nft_accuracies.append(nft_acc)

    # Calculate averages
    avg_bf_time = np.mean(bf_times)
    avg_nft_time = np.mean(nft_times)
    avg_bf_memory = np.mean(bf_memories)
    avg_nft_memory = np.mean(nft_memories)
    avg_bf_accuracy = np.mean(bf_accuracies)
    avg_nft_accuracy = np.mean(nft_accuracies)

    # Print final metrics
    print("Brute Force (BF) - Avg Time: {:.2f}s, Avg Memory: {:.2f} MB, Avg Accuracy: {:.2f}%".format(
        avg_bf_time, avg_bf_memory, avg_bf_accuracy * 100))
    print("Naive Fine-Tuning (NFT) - Avg Time: {:.2f}s, Avg Memory: {:.2f} MB, Avg Accuracy: {:.2f}%".format(
        avg_nft_time, avg_nft_memory, avg_nft_accuracy * 100))

    return bf_times, nft_times, bf_memories, nft_memories, bf_accuracies, nft_accuracies

def plot_results(bf_times, nft_times, bf_memories, nft_memories, bf_accuracies, nft_accuracies):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    ax1.plot(bf_times, label='Brute Force Time')
    ax1.plot(nft_times, label='Naive Fine-Tuning Time')
    ax1.set_title('Training Time per Batch')
    ax1.set_xlabel('Batch')
    ax1.set_ylabel('Time (seconds)')
    ax1.legend()
    
    ax2.plot(bf_accuracies, label='Brute Force Accuracy')
    ax2.plot(nft_accuracies, label='Naive Fine-Tuning Accuracy')
    ax2.set_title('Test Accuracy per Batch')
    ax2.set_xlabel('Batch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

# Run the experiment
if __name__ == "__main__":
    bf_times, nft_times, bf_memories, nft_memories, bf_accuracies, nft_accuracies = run_experiment()
    plot_results(bf_times, nft_times, bf_memories, nft_memories, bf_accuracies, nft_accuracies)
```

### Deucième objectif
Dans les deux approches BF et NFT, le modèle en entier est réentraıné a chaque arrivée de nouveau batch (soit avec l’ensemble de toutes les données disponibles pour BF soit avec le nouveau batch pour NFT).
La méthode Move-To-Data (MTD) propose une approche différente. Elle consiste à figer les premières couches du modèle et à ne re-entraıner que la toute dernière couche, celle permettant la classification.
En plus, l’approche MTD propose de ne pas utiliser la descente de gradient pour l’entraınement, mais un autre optimiseur plus adapté a l’apprentissage en ligne

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, Input
import matplotlib.pyplot as plt
import time
import psutil
import numpy as np

# Charger les données MNIST et Fashion-MNIST
def load_data_with_new_classes():
    (mnist_x_train, mnist_y_train), (mnist_x_test, mnist_y_test) = tf.keras.datasets.mnist.load_data()
    (fmnist_x_train, fmnist_y_train), (fmnist_x_test, fmnist_y_test) = tf.keras.datasets.fashion_mnist.load_data()

    mnist_x_train = mnist_x_train[..., np.newaxis].astype("float32") / 255.0
    mnist_x_test = mnist_x_test[..., np.newaxis].astype("float32") / 255.0
    fmnist_x_train = fmnist_x_train[..., np.newaxis].astype("float32") / 255.0
    fmnist_x_test = fmnist_x_test[..., np.newaxis].astype("float32") / 255.0

    fmnist_y_train = fmnist_y_train.copy() + 10
    fmnist_y_test = fmnist_y_test.copy() + 10

    x_initial = mnist_x_train[:5000]
    y_initial = mnist_y_train[:5000]
    x_new_classes = fmnist_x_train[:2000]
    y_new_classes = fmnist_y_train[:2000]

    x_test = np.concatenate([mnist_x_test[:1000], fmnist_x_test[:1000]])
    y_test = np.concatenate([mnist_y_test[:1000], fmnist_y_test[:1000]])

    return x_initial, y_initial, x_new_classes, y_new_classes, x_test, y_test

# Création du modèle de base
def create_base_model(num_classes=10):
    inputs = Input(shape=(28, 28, 1))
    x = layers.Conv2D(32, (3, 3), activation='relu')(inputs)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Conv2D(64, (3, 3), activation='relu')(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Conv2D(64, (3, 3), activation='relu')(x)
    x = layers.Flatten()(x)
    x = layers.Dense(128, activation='relu', name='features')(x)
    outputs = layers.Dense(num_classes, activation='softmax', name='classification')(x)
    model = models.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Entraînement initial
def train_initial_model(model, x_initial, y_initial):
    model.fit(x_initial, y_initial, epochs=5, batch_size=32, verbose=1)

# Mise à jour des poids avec MTD
def update_weights_mtd(model, x_new_classes, y_new_classes, epsilon=0.01):
    features_layer = models.Model(inputs=model.input, outputs=model.get_layer("features").output)
    weights, biases = model.get_layer("classification").get_weights()

    for i in range(len(x_new_classes)):
        x = x_new_classes[i:i+1]
        y = y_new_classes[i]
        h = features_layer.predict(x)
        h_norm = h / np.linalg.norm(h)
        weights[:, y] += epsilon * (np.linalg.norm(weights[:, y]) * h_norm[0] - weights[:, y])

    model.get_layer("classification").set_weights([weights, biases])

# Évaluation du modèle
def evaluate_model(model, x_test, y_test):
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    return accuracy

# Visualisation des résultats
def plot_results(batch_accuracies, batch_times, batch_memory_usage):
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.plot(batch_accuracies, marker='o')
    plt.title('Accuracy par Batch')
    plt.xlabel('Batch')
    plt.ylabel('Accuracy')

    plt.subplot(1, 3, 2)
    plt.plot(batch_times, marker='o')
    plt.title('Temps de calcul par Batch')
    plt.xlabel('Batch')
    plt.ylabel('Temps (s)')

    plt.subplot(1, 3, 3)
    plt.plot(batch_memory_usage, marker='o')
    plt.title('Mémoire utilisée par Batch')
    plt.xlabel('Batch')
    plt.ylabel('Mémoire (MB)')

    plt.tight_layout()
    plt.show()

    # Afficher les moyennes
    print(f"Accuracy moyenne: {np.mean(batch_accuracies) * 100:.2f}%")
    print(f"Temps de calcul moyen par batch: {np.mean(batch_times):.4f} secondes")
    print(f"Mémoire utilisée moyenne: {np.mean(batch_memory_usage):.2f} MB")
    
# Pipeline principal
def main():
    x_initial, y_initial, x_new_classes, y_new_classes, x_test, y_test = load_data_with_new_classes()

    model = create_base_model(num_classes=20)
    train_initial_model(model, x_initial, y_initial)

    batch_size = 100
    batch_accuracies = []
    batch_times = []
    batch_memory_usage = []

    for i in range(0, len(x_new_classes), batch_size):
        batch_x = x_new_classes[i:i + batch_size]
        batch_y = y_new_classes[i:i + batch_size]

        # Mesure du temps
        start_time = time.time()
        
        # Mesure de la mémoire avant
        process = psutil.Process()
        memory_before = process.memory_info().rss / 1024 / 1024  # Conversion en MB
        
        update_weights_mtd(model, batch_x, batch_y)
        
        # Mesure de la mémoire après
        memory_after = process.memory_info().rss / 1024 / 1024
        memory_used = memory_after - memory_before
        
        end_time = time.time()

        accuracy = evaluate_model(model, x_test, y_test)
        batch_accuracies.append(accuracy)
        batch_times.append(end_time - start_time)
        batch_memory_usage.append(memory_used)

    plot_results(batch_accuracies, batch_times, batch_memory_usage)

if __name__ == "__main__":
    main()
```

### Troisième objectif apprentissage avec de nouvelles classes
Dans les section précédantes, nous avons considéré que les batchs de données arrivant appartenaaient à la meme classe que les données d’entraınement. Dans cette section, nous allons considérer le cas ou les nouvelles données appartiennent à  de nouvelles classes. Quelles adaptations sont nécessaires pour les méthodes vues précédamment? Comment les performances sont-elles impactées?

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, Input
import matplotlib.pyplot as plt
import time
import tracemalloc
import seaborn as sns

def load_data_sequential_classes():
    # Charger MNIST et Fashion-MNIST
    (mnist_x_train, mnist_y_train), (mnist_x_test, mnist_y_test) = tf.keras.datasets.mnist.load_data()
    (fmnist_x_train, fmnist_y_train), (fmnist_x_test, fmnist_y_test) = tf.keras.datasets.fashion_mnist.load_data()
    
    # Normalisation
    mnist_x_train = mnist_x_train[..., np.newaxis].astype("float32") / 255.0
    mnist_x_test = mnist_x_test[..., np.newaxis].astype("float32") / 255.0
    fmnist_x_train = fmnist_x_train[..., np.newaxis].astype("float32") / 255.0
    fmnist_x_test = fmnist_x_test[..., np.newaxis].astype("float32") / 255.0
    
    # Préparation des données initiales (5 classes de MNIST)
    initial_classes = [0, 1, 2, 3, 4]
    new_classes_batch1 = [5, 6, 7, 8, 9]  # Reste des classes MNIST
    new_classes_batch2 = [10, 11, 12, 13, 14]  # Premières classes Fashion-MNIST
    new_classes_batch3 = [15, 16, 17, 18, 19]  # Reste des classes Fashion-MNIST
    
    # Données initiales
    initial_mask = np.isin(mnist_y_train, initial_classes)
    x_initial = mnist_x_train[initial_mask]
    y_initial = mnist_y_train[initial_mask]
    
    # Premier batch de nouvelles classes
    batch1_mask = np.isin(mnist_y_train, new_classes_batch1)
    x_batch1 = mnist_x_train[batch1_mask]
    y_batch1 = mnist_y_train[batch1_mask]
    
    # Deuxième batch de nouvelles classes (Fashion-MNIST)
    batch2_mask = np.isin(fmnist_y_train, [0, 1, 2, 3, 4])
    x_batch2 = fmnist_x_train[batch2_mask]
    y_batch2 = fmnist_y_train[batch2_mask] + 10
    
    # Troisième batch de nouvelles classes (Fashion-MNIST)
    batch3_mask = np.isin(fmnist_y_train, [5, 6, 7, 8, 9])
    x_batch3 = fmnist_x_train[batch3_mask]
    y_batch3 = fmnist_y_train[batch3_mask] + 10
    
    # Données de test
    test_mask_mnist = np.isin(mnist_y_test, range(10))
    test_mask_fmnist = np.isin(fmnist_y_test, range(10))
    
    x_test = np.concatenate([
        mnist_x_test[test_mask_mnist], 
        fmnist_x_test[test_mask_fmnist]
    ])
    y_test = np.concatenate([
        mnist_y_test[test_mask_mnist],
        fmnist_y_test[test_mask_fmnist] + 10
    ])
    
    return (x_initial, y_initial), [(x_batch1, y_batch1), (x_batch2, y_batch2), (x_batch3, y_batch3)], (x_test, y_test)

def create_expandable_model(initial_classes=5):
    inputs = Input(shape=(28, 28, 1))
    x = layers.Conv2D(32, (3, 3), activation='relu')(inputs)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Conv2D(64, (3, 3), activation='relu')(x)
    x = layers.MaxPooling2D((2, 2))(x)
    x = layers.Conv2D(64, (3, 3), activation='relu')(x)
    x = layers.Flatten()(x)
    features = layers.Dense(128, activation='relu', name='features')(x)
    outputs = layers.Dense(20, activation='softmax', name='classification')(features)
    
    model = models.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

def brute_force_new_classes(model, new_batch_x, new_batch_y, all_previous_x, all_previous_y):
    tracemalloc.start()
    start_time = time.time()
    
    x_combined = np.concatenate([all_previous_x, new_batch_x])
    y_combined = np.concatenate([all_previous_y, new_batch_y])
    
    history = model.fit(x_combined, y_combined, epochs=5, batch_size=32, verbose=0)
    
    training_time = time.time() - start_time
    _, peak_memory = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    
    return model, training_time, peak_memory / 1e6, history.history['accuracy'][-1]

def nft_new_classes(model, new_batch_x, new_batch_y):
    tracemalloc.start()
    start_time = time.time()
    
    history = model.fit(new_batch_x, new_batch_y, epochs=5, batch_size=32, verbose=0)
    
    training_time = time.time() - start_time
    _, peak_memory = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    
    return model, training_time, peak_memory / 1e6, history.history['accuracy'][-1]

def mtd_new_classes(model, new_batch_x, new_batch_y, epsilon=0.1, batch_size=32, iterations=3):
    tracemalloc.start()
    start_time = time.time()
    
    features_layer = models.Model(inputs=model.input, outputs=model.get_layer("features").output)
    weights, biases = model.get_layer("classification").get_weights()
    
    # Multiple iterations over the dataset
    for iteration in range(iterations):
        # Shuffle data
        indices = np.random.permutation(len(new_batch_x))
        new_batch_x = new_batch_x[indices]
        new_batch_y = new_batch_y[indices]
        
        for i in range(0, len(new_batch_x), batch_size):
            batch_end = min(i + batch_size, len(new_batch_x))
            x_batch = new_batch_x[i:batch_end]
            y_batch = new_batch_y[i:batch_end]
            
            # Get features
            h = features_layer.predict(x_batch, verbose=0)
            
            # Adaptive epsilon based on iteration
            current_epsilon = epsilon * (1 - 0.2 * iteration)
            
            for j in range(len(x_batch)):
                h_norm = h[j] / np.linalg.norm(h[j])
                
                # Update weights with momentum
                momentum = 0.9
                old_weights = weights[:, y_batch[j]].copy()
                weights[:, y_batch[j]] += current_epsilon * (
                    np.linalg.norm(weights[:, y_batch[j]]) * h_norm - 
                    weights[:, y_batch[j]]
                )
                weights[:, y_batch[j]] = momentum * old_weights + (1 - momentum) * weights[:, y_batch[j]]
    
    model.get_layer("classification").set_weights([weights, biases])
    
    training_time = time.time() - start_time
    _, peak_memory = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    
    return model, training_time, peak_memory / 1e6

def evaluate_per_class(model, x_test, y_test):
    predictions = model.predict(x_test, verbose=0)
    pred_classes = np.argmax(predictions, axis=1)
    
    class_accuracies = {}
    for class_idx in range(20):
        mask = y_test == class_idx
        if np.any(mask):
            class_acc = np.mean(pred_classes[mask] == y_test[mask])
            class_accuracies[class_idx] = class_acc
    
    return class_accuracies

def plot_comparative_results(bf_metrics, nft_metrics, mtd_metrics):
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Définir les numéros des batchs
    batches = range(1, len(bf_metrics['accuracies']) + 1)
    
    # Graphique de l'évolution de la précision
    axes[0, 0].plot(batches, bf_metrics['accuracies'], label='Brute Force', marker='o')
    axes[0, 0].plot(batches, nft_metrics['accuracies'], label='NFT', marker='s')
    axes[0, 0].plot(batches, mtd_metrics['accuracies'], label='MTD', marker='^')
    axes[0, 0].set_title('Accuracy Evolution')
    axes[0, 0].set_xlabel('Batch')
    axes[0, 0].set_ylabel('Accuracy')
    axes[0, 0].set_xticks(batches)
    axes[0, 0].legend()
    
    # Graphique du temps d'entraînement par batch
    axes[0, 1].plot(batches, bf_metrics['times'], label='Brute Force', marker='o')
    axes[0, 1].plot(batches, nft_metrics['times'], label='NFT', marker='s')
    axes[0, 1].plot(batches, mtd_metrics['times'], label='MTD', marker='^')
    axes[0, 1].set_title('Training Time per Batch')
    axes[0, 1].set_xlabel('Batch')
    axes[0, 1].set_ylabel('Time (seconds)')
    axes[0, 1].set_xticks(batches)
    axes[0, 1].legend()
    
    # Graphique de la mémoire utilisée
    axes[1, 0].plot(batches, bf_metrics['memories'], label='Brute Force', marker='o')
    axes[1, 0].plot(batches, nft_metrics['memories'], label='NFT', marker='s')
    axes[1, 0].plot(batches, mtd_metrics['memories'], label='MTD', marker='^')
    axes[1, 0].set_title('Peak Memory per Batch')
    axes[1, 0].set_xlabel('Batch')
    axes[1, 0].set_ylabel('Memory (MB)')
    axes[1, 0].set_xticks(batches)
    axes[1, 0].legend()
    
    # Suppression du quatrième subplot vide
    fig.delaxes(axes[1, 1])
    
    plt.tight_layout()
    plt.show()


def run_experiment():
    # Chargement des données
    (x_initial, y_initial), new_class_batches, (x_test, y_test) = load_data_sequential_classes()
    
    # Initialisation des modèles
    bf_model = create_expandable_model()
    nft_model = create_expandable_model()
    mtd_model = create_expandable_model()
    
    # Entraînement initial
    print("Entraînement initial...")
    initial_history = bf_model.fit(x_initial, y_initial, epochs=5, batch_size=32, verbose=1)
    nft_model.set_weights(bf_model.get_weights())
    mtd_model.set_weights(bf_model.get_weights())
    
    # Métriques
    bf_metrics = {'times': [], 'memories': [], 'accuracies': [], 'class_accuracies': []}
    nft_metrics = {'times': [], 'memories': [], 'accuracies': [], 'class_accuracies': []}
    mtd_metrics = {'times': [], 'memories': [], 'accuracies': [], 'class_accuracies': []}
    
    prev_x = x_initial.copy()
    prev_y = y_initial.copy()
    
    # Apprentissage incrémental
    print("\nApprentissage des nouvelles classes...")
    for batch_idx, (batch_x, batch_y) in enumerate(new_class_batches):
        print(f"\nTraitement du batch {batch_idx + 1}...")
        
        # Brute Force
        bf_model, bf_time, bf_memory, bf_acc = brute_force_new_classes(
            bf_model, batch_x, batch_y, prev_x, prev_y)
        bf_class_acc = evaluate_per_class(bf_model, x_test, y_test)
        
        # NFT
        nft_model, nft_time, nft_memory, nft_acc = nft_new_classes(
            nft_model, batch_x, batch_y)
        nft_class_acc = evaluate_per_class(nft_model, x_test, y_test)
        
        # MTD
        mtd_model, mtd_time, mtd_memory = mtd_new_classes(
            mtd_model, batch_x, batch_y)
        _, mtd_acc = mtd_model.evaluate(x_test, y_test, verbose=0)
        mtd_class_acc = evaluate_per_class(mtd_model, x_test, y_test)
        
        # Mise à jour des métriques
        for metrics, time, memory, acc, class_acc in [
            (bf_metrics, bf_time, bf_memory, bf_acc, bf_class_acc),
            (nft_metrics, nft_time, nft_memory, nft_acc, nft_class_acc),
            (mtd_metrics, mtd_time, mtd_memory, mtd_acc, mtd_class_acc)
        ]:
            metrics['times'].append(time)
            metrics['memories'].append(memory)
            metrics['accuracies'].append(acc)
            metrics['class_accuracies'].append(class_acc)
        
        prev_x = np.concatenate([prev_x, batch_x])
        prev_y = np.concatenate([prev_y, batch_y])
    
    # Affichage des résultats
    print("\nRésultats finaux:")
    for name, metrics in [("Brute Force", bf_metrics), ("NFT", nft_metrics), ("MTD", mtd_metrics)]:
        print(f"\n{name}:")
        print(f"Temps moyen: {np.mean(metrics['times']):.2f}s")
        print(f"Mémoire moyenne: {np.mean(metrics['memories']):.2f}MB")
        print(f"Accuracy finale: {metrics['accuracies'][-1]*100:.2f}%")
    
    return bf_metrics, nft_metrics, mtd_metrics

if __name__ == "__main__":
    bf_metrics, nft_metrics, mtd_metrics = run_experiment()
    plot_comparative_results(bf_metrics, nft_metrics, mtd_metrics)
```
